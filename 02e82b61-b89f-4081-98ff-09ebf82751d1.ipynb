2. C подпиской Ultra:

Создадим функции для упрощения и удобства. 

```python
#вывод основной информации о датафрейме
def data_info(df):
    display('Первые строки')
    display(df.head())
    display('Описание данных')
    display(df.describe())
    display('Базовая системная информация')
    display(df.info())
```

```python
#функция для вывода пропусков и дубликатов в количественном и процентном виде
def mis_dup_info(df):
    missing_values = df.isnull().sum() # количество пропущеннных строк
    duplicated_values = df.duplicated().sum() #Количество явных дубликатов
    total_values = df.shape[0]  # общее количество строк

    # добавление процента значений
    missing_percent = ((missing_values / total_values) * 100).round(1).astype(str) + '%'
    duplicated_percent = ((duplicated_values / total_values) * 100).round(1).astype(str) + '%'
    
    # создание датафреймов отображающих нужные данные
    missing_duplicated_df = pd.DataFrame({
        'total' : total_values,
        'missing': missing_values,
        'percent_missing': missing_percent,
        'duplicated' : duplicated_values,
        'percent_duplicated' : duplicated_percent
        
    })   
    

    return missing_duplicated_df
```

## Шаг 5. Подсчёт выручки

### Создание датафрейма с агрегированными данными о поездках

Создадим датафрейм с агрегированными данными о поездках на основе датафрейма с объединёнными данными из шага 4: найдем суммарное расстояние, количество поездок и суммарное время для каждого пользователя за каждый месяц.

<font color='DarkBlue'><b>Комментарий ревьюера</b></font><br>
<font color='DarkGreen'>👌 Правильно сделано округление продолжительности поездки.</font>

```python
#округляем к верхнему значению
merged_data['duration'] = np.ceil(rides_data['duration']) 
#меняем тип данных duration на int
merged_data['duration'] = merged_data['duration'].astype('int')

merged_pivot = merged_data.pivot_table(
    index=['user_id', 'month'],
    values=['distance', 'duration'],
    aggfunc=['sum', 'count']
)
#для того чтобы столбцы выглядели более компактно и удобно
merged_pivot.columns = merged_pivot.columns.map(lambda col: '_'.join(col))
merged_pivot.drop('count_duration', axis=1, inplace=True)
merged_pivot.rename(columns={"count_distance" : "count"}, inplace=True)

merged_pivot
```

### Добавим столбец с помесячной выручкой, которую принёс каждый пользователь.

Помесячная выручка считается по формуле: стоимость старта поездки * количество поездок + стоимость одной минуты поездки * общая продолжительность всех поездок в минутах + стоимость подписки.  
Создадим функцию:

```python
def calc(srs):
    x = srs['start_ride_price'] * srs['count'] + srs['minute_price'] * srs['sum_duration'] + srs['subscription_fee']
    return x
```

Сделаем вычисление.

```python
merged_pivot['income'] = merged_pivot.apply(calc, axis=1)
merged_pivot
```

То есть по завершении периода действия подписки пользователь может либо отказаться от неё, либо продлить, совершив соответствующий платёж.  

Эта акция уже проводилась ранее и по итогу выяснилось, что после бесплатного пробного периода подписку продлевают 10 % пользователей. Выясним, какое минимальное количество промокодов нужно разослать, чтобы вероятность не выполнить план была примерно 5 %. Подберем параметры распределения, описывающие эту ситуацию, пострем график распределения и сформулируем ответ на вопрос о количестве промокодов.

```python
p = 0.1  # вероятность продления подписки
a = 100  # нужно как минимум 100 подписок
alpha = 0.05  # допустимая вероятность невыполнения плана

# подбираем минимальное n
n = a  # начнем со 100 и будем увеличивать
while st.binom.cdf(a - 1, n, p) > alpha:
    n += 1

# визуализация распределения
x = np.arange(a - 20, a + 50)  # ось x
pmf_values = st.binom.pmf(x, n, p)  # вычисление вероятностей

# строим график
plt.figure(figsize=(10, 5))
plt.plot(x, pmf_values)
plt.xlabel("Количество продлений подписки")
plt.ylabel("Вероятность")
plt.title("Распределение вероятности продления подписки")
plt.grid()

# создаем линию
plt.axvline(a, color='red')

plt.show()

print(f"Минимальное количество разосланных промокодов: {n}")
```

### Оценка вероятности открытия push-уведомлений с нормальной аппроксимацией

Отдел маркетинга рассылает клиентам push-уведомления в мобильном приложении. Клиенты могут открыть его или не открывать. Известно, что уведомления открывают около 40 % получивших клиентов. Отдел планирует разослать 1 млн уведомлений. С помощью аппроксимации построим примерный график распределения и оценим вероятность того, что уведомление откроют не более 399,5 тыс. пользователей.

Эта ситуация тоже описывается биномиальным распределением. Но считать каждое отдельное значение достаточно долго.  воспользуемся нормальной аппроксимацией биномиального распределения и cdf() для быстрой оценки.

```python
n = 1000000  # количество разосланных уведомлений
p = 0.4  # вероятность открытия
a = 399500  # нужное значение

# вычисление параметров нормального распределения
mu = n * p  # среднее
sigma = np.sqrt(n * p * (1 - p))  # стандартное отклонение

# создание нормального распределения
distr = st.norm(mu, sigma)

# вычисление вероятности 
result = distr.cdf(a)

# визуализация распределения
x = np.linspace(mu - 4 * sigma, mu + 4 * sigma, 1000) # ось x
y = distr.pdf(x)  # функция плотности вероятности (pdf), не pmf потому-что функция неприрывная 

plt.figure(figsize=(10, 5))
plt.plot(x, y)
plt.xlabel("Количество открытий")
plt.ylabel("Плотность вероятности")
plt.title("Распределение вероятности открытия уведломления")
plt.grid()

#вертикальная линия для a = 399500
plt.axvline(a, color='red')

# отображение графика
plt.show()

print(f'Вероятность того что уведомление откроют не более 399500 раз - {result.round(4)}%')
```

Отдел маркетинга может рассчитывать, что с высокой вероятностью (84%) количество открытых уведомлений превысит 399500, а вероятность того, что оно окажется ниже, составляет 15.37%.

Несмотря на различия в количестве данных, медианные значения продолжительности поездок для обеих групп схожи, однако у пользователей без подписки немного преобладают короткие поездки, а у подписчиков — более длинные.  

После, был произведен подсчет выручки, для этого мы создали датафрейм с агрегированными данными о поездках, включающий в себя колонки о суммарном количестве, длительности, и расстоянию поездок по месяцам и пользователям, на основе этого, по формуле мы произвели расчеты с помощью функции. 

В следущей главе мы перешли к проверке гипотез. 
Мы узнали, что:  
Cредняя продолжительность поездки у подписчиков больше, чем у пользователей без подписки, наличие подписки увеличивает доходность пользователей, что может свидетельствовать о том, что подписчики тратят больше средств.  
Среднее время пользования самокатом не привышает оптимальный порог с точки зрения износа самоката в 3130 метров, так же расписали тест на изменение количества обращений в техподдержку после обновления серверов. 

В следущей главе мы приступили к решениям заданий по распределению, определили минимальное количество промокодов для достижения целевого продления подписки и пришли к выводу, чтобы с 95% вероятностью как минимум 100 пользователей продлили подписку после получения промокодов, необходимо разослать не менее 1161 промокода.  
Сделали оценку вероятности открытия push-уведомлений, при рассылке 1 миллиона push-уведомлений и средней доле открытий в 40% вероятность того, что менее 399,5 тысячи пользователей откроют уведомление, составляет всего 0,1537%. Это довольно низкое значение, что означает высокую вероятность (84%) того, что число открытых уведомлений превысит данный порог.
